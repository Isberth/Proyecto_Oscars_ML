```python
# 1Ô∏è‚É£ Cargar librer√≠as necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import GradientBoostingClassifier

# 2Ô∏è‚É£ Cargar el dataset
ruta_archivo = "/content/drive/MyDrive/Colab Notebooks/Proyecto M7/Oscars-demographics-DFE.csv"
df = pd.read_csv(ruta_archivo, encoding="latin1")

# 3Ô∏è‚É£ Eliminar columnas irrelevantes
columnas_a_eliminar = [
    "_unit_id", "_golden", "_trusted_judgments", "_unit_state", "biourl",
    "birthplace_gold", "date_of_birth_gold", "race_ethnicity_gold",
    "religion_gold", "sexual_orientation_gold", "year_of_award_gold",
    "birthplace:confidence", "date_of_birth:confidence", "race_ethnicity:confidence",
    "religion:confidence", "sexual_orientation:confidence", "year_of_award:confidence",
    "_last_judgment_at"
]
df_clean = df.drop(columns=columnas_a_eliminar, errors='ignore')

print("\nüîπ Columnas eliminadas. Dimensiones actuales del dataset:", df_clean.shape)

# 4Ô∏è‚É£ Convertir `date_of_birth` a formato de fecha (asegurando que sea datetime)
df_clean["date_of_birth"] = pd.to_datetime(df_clean["date_of_birth"], errors="coerce")

# Revisar si hay valores nulos despu√©s de la conversi√≥n
if df_clean["date_of_birth"].isnull().sum() > 0:
    print(f"‚ö† Advertencia: {df_clean['date_of_birth'].isnull().sum()} valores nulos en `date_of_birth` tras conversi√≥n.")

# 5Ô∏è‚É£ Corregir `country` extrayendo el √∫ltimo valor despu√©s de la coma
df_clean["country"] = df_clean["birthplace"].str.split(',').str[-1].str.strip()

# 6Ô∏è‚É£ Eliminar valores incorrectos en `country`
correcciones_paises = {"Il": "United States", "Ut": "United States"}
df_clean["country"] = df_clean["country"].replace(correcciones_paises)

# 7Ô∏è‚É£ Crear variable de edad al ganar el Oscar
df_clean["age_at_award"] = df_clean["year_of_award"] - df_clean["date_of_birth"].dt.year

# 8Ô∏è‚É£ Eliminar valores nulos en `age_at_award`
df_clean.dropna(subset=['age_at_award'], inplace=True)

# 9Ô∏è‚É£ Normalizar variables categ√≥ricas
df_clean[["race_ethnicity", "religion", "sexual_orientation"]] = df_clean[["race_ethnicity", "religion", "sexual_orientation"]].apply(lambda x: x.str.lower().str.strip())

# üîü Reemplazar valores inconsistentes en religi√≥n y orientaci√≥n sexual
df_clean[["religion", "sexual_orientation"]] = df_clean[["religion", "sexual_orientation"]].replace({
    "na": "desconocido",
    "see note": "desconocido",
    "matter of dispute": "desconocido"
})

# üìä **Gr√°ficos del An√°lisis Exploratorio**
# 1Ô∏è‚É£ Distribuci√≥n de la Edad de los Ganadores
plt.figure(figsize=(12, 6))
sns.histplot(df_clean['age_at_award'], bins=30, kde=True, color="blue", alpha=0.6)
plt.title("Distribuci√≥n de la Edad de los Ganadores del Oscar", fontsize=14)
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()

# 2Ô∏è‚É£ Distribuci√≥n de Premios por A√±o
plt.figure(figsize=(12, 6))
sns.histplot(df_clean['year_of_award'], bins=30, kde=True, color="green", alpha=0.6)
plt.title("Distribuci√≥n de A√±os de los Premios Oscar", fontsize=14)
plt.xlabel("A√±o del Premio")
plt.ylabel("Frecuencia")
plt.show()

# 3Ô∏è‚É£ Boxplot de Edad por Categor√≠a de Premio
plt.figure(figsize=(12, 6))
sns.boxplot(x="award", y="age_at_award", data=df_clean, hue="award", legend=False, palette="viridis")
plt.xticks(rotation=90)
plt.title("Distribuci√≥n de Edad por Categor√≠a de Premio", fontsize=14)
plt.xlabel("Categor√≠a del Premio")
plt.ylabel("Edad")
plt.show()

# 4Ô∏è‚É£ Gr√°fico de barras de los pa√≠ses con m√°s ganadores
plt.figure(figsize=(12, 6))
top_countries = df_clean['country'].value_counts().nlargest(10)
sns.barplot(x=top_countries.index, y=top_countries.values, hue=top_countries.index, legend=False, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Top 10 Pa√≠ses con M√°s Ganadores de Oscar", fontsize=14)
plt.xlabel("Pa√≠s")
plt.ylabel("Cantidad de Ganadores")
plt.show()

# 5Ô∏è‚É£ Mapa de Calor de Correlaciones
plt.figure(figsize=(10, 6))
df_numeric = df_clean.select_dtypes(include=['number'])  # Filtrar solo variables num√©ricas
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Calor de Correlaciones", fontsize=14)
plt.show()

# üîπ Confirmaci√≥n final
print("\nüîπ Informaci√≥n despu√©s de la limpieza final:")
df_clean.info()

print("\nüîπ Primeras filas despu√©s de la limpieza final:")
display(df_clean.head())

```

```python
# üîπ Paso siguiente: Preparaci√≥n de los datos para entrenamiento

# Importar librer√≠as adicionales necesarias
from sklearn.preprocessing import LabelEncoder

# Seleccionar variables relevantes para el modelo
df_model = df_clean.copy()

# Variables a utilizar como predictores (features)
variables_a_usar = [
    'age_at_award',
    'country',
    'race_ethnicity',
    'religion',
    'sexual_orientation'
]

# Variable objetivo
variable_objetivo = 'award'

# Crear DataFrame de features (X) y objetivo (y)
X = df_model[variables_a_usar]
y = df_model[variable_objetivo]

# üî∏ Codificaci√≥n de variables categ√≥ricas con LabelEncoder
label_encoders = {}
for columna in ['country', 'race_ethnicity', 'religion', 'sexual_orientation']:
    le = LabelEncoder()
    df_model[columna+'_encoded'] = le.fit_transform(df_model[columna])
    variables_a_usar.remove(columna)
    variables_a_usar.append(columna+'_encoded')

# DataFrame final para el modelado
X_final = df_model[variables_a_usar]
y = df_model[variable_objetivo]

# Divisi√≥n del dataset (80% entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(
    df_model[variables_a_usar],
    df_model[variable_objetivo],
    test_size=0.2,
    random_state=42,
    stratify=df_model[variable_objetivo]  # para mantener proporciones de categor√≠as
)

# Confirmaci√≥n
print("\nüîπ Forma del conjunto de entrenamiento:", X_train.shape, y_train.shape)
print("üîπ Forma del conjunto de prueba:", X_test.shape, y_test.shape)
```

```python
# üîπ Modelo inicial Gradient Boosting Classifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report

modelo_gb = GradientBoostingClassifier(random_state=42)

# Entrenamiento del modelo
modelo_gb.fit(X_train, y_train)

# Predicciones
predictions = modelo_gb.predict(X_test)

# Reporte de clasificaci√≥n
print("üîπ Classification Report del Modelo Gradient Boosting:")
print(classification_report(y_test, predictions))

# Accuracy del modelo
accuracy = accuracy_score(y_test, predictions)
print("üî∏ Accuracy del modelo:", accuracy)
```
